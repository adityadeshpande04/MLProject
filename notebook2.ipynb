```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Variant_III.csv")
df.head(5)
```

```python
fraud=len(df[df['fraud_bool']==0])
no_fraud=len(df)-fraud

print(no_fraud)
print(fraud)
```

```python
df.drop('prev_address_months_count', axis=1, inplace=True)
df.drop('intended_balcon_amount', axis=1, inplace=True)
df.drop('device_os', axis=1, inplace=True)
df.drop('zip_count_4w', axis=1, inplace=True)
df.drop('velocity_4w', axis=1, inplace=True)
df.drop('velocity_24h', axis=1, inplace=True)
df.drop('velocity_6h', axis=1, inplace=True)
df.drop('bank_branch_count_8w', axis=1, inplace=True)
df.drop('date_of_birth_distinct_emails_4w', axis=1, inplace=True)
```

```python
low_income=0.2
high_income=0.7

def category_income(income):
    if income <= low_income:
        return 1
    elif low_income < income <= high_income:
        return 2
    else:
        return 3

df['income']=df['income'].apply(category_income)
df['income']=df['income'].astype('int64')

df.head(10)
    
```

```python
df['name_email_similarity']=df['name_email_similarity'].apply(lambda x: round(x,3))

low=0.3
high=0.6
def category_email(similarity):
    if similarity <low:
        return 1
    elif low < similarity <= high:
        return 2
    else:
        return 3

df['name_email_similarity']=df['name_email_similarity'].apply(category_email)
df['name_email_similarity']=df['name_email_similarity'].astype('int64')

df.head(10)
```

```python
num_bins = 5  
column_to_partition = 'credit_risk_score'
min_credit_score = df[column_to_partition].min()-1

bin_width = (df[column_to_partition].max() - min_credit_score) / num_bins

bin_edges = [min_credit_score + i * bin_width for i in range(num_bins + 1)]

bin_labels = [i for i in range(1, num_bins + 1)]

df['credit_risk_score'] = pd.cut(df[column_to_partition], bins=bin_edges, labels=bin_labels)
df['credit_risk_score'] = df['credit_risk_score'].astype('int64')
print(df['credit_risk_score'])
```

```python


df = pd.get_dummies(df, columns=['payment_type'], prefix='pt', drop_first=False)
os_columns = [col for col in df.columns if col.startswith('pt_')]

df = pd.get_dummies(df, columns=['employment_status'], prefix='emp', drop_first=False)
os_columns = [col for col in df.columns if col.startswith('emp_')]

df = pd.get_dummies(df, columns=['housing_status'], prefix='hs', drop_first=False)
os_columns = [col for col in df.columns if col.startswith('hs_')]

df = pd.get_dummies(df, columns=['source'], prefix='src', drop_first=False)
os_columns = [col for col in df.columns if col.startswith('src_')]

df.replace({False: 0, True: 1}, inplace=True)


```

```python
df.dropna(axis=0, inplace=True)
```

```python
majority_class = df[df['fraud_bool'] == 0]
minority_class=df[df['fraud_bool']==1]

desired_majority_samples=70000
desired_minority_samples=500
undersampled_majority = majority_class.sample(n=desired_majority_samples, random_state=42)
undersampled_minority = minority_class.sample(n=desired_minority_samples, random_state=42)
undersampled_df = pd.concat([undersampled_majority,undersampled_minority])
undersampled_df = undersampled_df.sample(frac=1, random_state=42).reset_index(drop=True)
undersampled_y=undersampled_df['fraud_bool']
undersampled_x=undersampled_df.drop(columns=['fraud_bool'])

# X_train=undersampled_majority.drop(columns=['fraud_bool'])
# X_test= undersampled_df.drop(columns=['fraud_bool'])
# y_test=undersampled_df['fraud_bool']

from sklearn import svm
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(undersampled_x, undersampled_y, test_size=0.3, random_state=42)

model = svm.OneClassSVM(kernel='rbf', gamma=0.1)
model.fit(X_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)




```

```python
from sklearn.neighbors import NearestNeighbors

k=3
threshold=0.5

knn_model=NearestNeighbors(n_neighbors=k)
knn_model.fit(X_train,y_train)

distances, _ = knn_model.kneighbors(X_test)

kth_distances = distances[:, -1]

y_pred=[]

for distance in kth_distances:
    if distance > threshold:
        y_pred.append(0)
    else:
        y_pred.append(1)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.ensemble import IsolationForest

isolation_forest_model = IsolationForest(contamination= 0.01 , random_state=42)

isolation_forest_model.fit(X_train)

y_pred = isolation_forest_model.predict(X_test)

# Convert predictions to binary labels: 1 for inliers (normal), 0 for outliers (anomalies)
y_pred_binary = [1 if x == 1 else 0 for x in y_pred]

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print("Accuracy:", accuracy)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_binary)
print("Confusion Matrix:")
print(conf_matrix)
```
